
# Q5. Select Data using loc

## Problem Statement
Create a DataFrame with Name, Age and City.

Using `loc`, display the Name and City of the first row.

## Sample Output
```python
Name    Aman
City    Bhopal
```


````python
import pandas as pd
data = {
    "Name" : ["Aman","Rahul","Abhishek"],
    "Age" : [21,22,23],
    "city" : ["Bhopal","Mumbai","Patna"]

}
df = pd.DataFrame(data)
# .loc --> It is used to select rows and columns by their labels

 #   df.loc[ column_label].
print(df.loc[1,["Name","city"]])
````


# Q6. Select Data using iloc

## Problem Statement
Create a DataFrame with at least 4 rows.

Using `iloc`, print:
- Second row
- First two columns

## Sample Output
```python
Selected row values
```


````python
# loc → Uses row/column labels (names).
# iloc → Uses row/column positions (numbers).
import pandas as pd
data = {
    "Name" : ["Rahul","Mohit","Abhijeet","Ayush Tiwari"],
    "Age" : [23,38,48,29],
    "city": ["Patna","chapra","Delhi","Mumbai"]

}
df = pd.DataFrame(data)
print(df.iloc[1])  # Row 1 ka .iloc [sab print kar dega]
# yedi mai kisi rows ka sirf Age ya sirf City print karana chau to
# print(df.iloc[1]["Age"])

print(df.iloc[0:3,0:2]) # [Rows ,column] # isme hum chahte hai ki sirf rows me


# conclusion - {.iloc }--> ka use rows aur colums ko index wise print karana ho
# {.loc[location]} --> ka use columns lo name wise print karne ke liye hota hai

````


# Q7. Rename a Column

## Problem Statement
Create a DataFrame:

```python
Name
Marks
```

Rename column `Marks` to `Score`.

## Sample Output
```python
Name  Score
```


````python
import pandas as pd
data = {
    "Name" : ["Aman","pankaj","Ravi"],
    "Marks" : [80,75,80]
}
df = pd.DataFrame(data)
df = df.rename(columns = {"Marks" : "Score"})  # df.rename column me present
# Marks ko score me convert kar dega
print(df)
````


# Q8. Handle Missing Values

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman", None, "Neha"]
Marks = [80, 75, None]
```

Fill missing values using `fillna()`.

## Sample Output
```python
Missing values replaced
```


````python
data = {
    "Name" : ["Aman",None,"Neha"],
    "Marks" : [80,75,None]
}
df = pd.DataFrame(data)
df["Name"].fillna("Abhishek",inplace = True)  # fillna pandas ka ek function hai
# jo missing value/None value ko fill karne ke liye use kiya jata hai
df["Marks"].fillna(78 ,inplace=True)
print(df)
````


# Q9. Replace Values

## Problem Statement
Create a DataFrame:

```python
City = ["BPL", "IND", "BPL"]
```

Replace:
- BPL → Bhopal
- IND → Indore

## Sample Output
```python
Bhopal
Indore
Bhopal
```


# Q10. Remove Duplicate Records

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman","Rahul","Aman","Neha"]
```

Check duplicate rows using `duplicated()` and remove them using `drop_duplicates()`.

## Sample Output
```python
Aman
Rahul
Neha
```


````python
data = {
    "Name" : ["Aman","Rahul","Aman","Neha"]
}
df = pd.DataFrame(data)
#df = df.duplicated() --> duplicated() -- Duplicate rows identify
 # ko identify karta hai (True/False)

df = df.drop_duplicates()  # drop_duplicates() --> duplicate rows ko remove
# karta hai
print(df)
````
