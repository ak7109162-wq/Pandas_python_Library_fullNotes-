## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
import pandas as pd
marks = pd.Series([30,40,50,60,70],
                  index = ["Ronaldo","Messi","Virat","Dhoni","Mbappe"])
marks
````

### Creating a DataFrame

````python
info = pd.DataFrame(  # Many column in a frame iss called dataframe
    {
    "name" : ["Rahul","Deveraj","Abhishek","Rishi"],
    "Age" : [20,21,22,20],
    "marks" : [100,80,70,85]

})
print(info)

# selecting a column
print(info["name"])

# To get specific row we use .loc
print(info.loc)

# select a specific value
print(info.loc[1,"name"])
````

### Creating DataFrame from NumPy Arrays

````python
information = pd.DataFrame([
    ["Rahul",52,80],
    ["Deveraj",22,85],
    ["Abhishek",60,55]
    ], columns = ["Name","Age","Marks"])
print(information)
print(information.loc[2])
print(information.loc[2,"Marks"])
````

````python
# Generate a 3 x 4 matrix with random numbers(1-13) using numpy
#name the columns as [A,B,C,D]
import numpy as np
arr = np.random.randint(1,100,size =(3,4))
arr_df = pd.DataFrame(arr,columns=["A","B","C","D"])
print(arr_df)
````
