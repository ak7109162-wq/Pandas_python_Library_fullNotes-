## Handling Missing Values
Real-world datasets often have missing data. Pandas offers:
- `isna()` / `isnull()` → detect missing values (returns True/False)
- `dropna()` → remove missing values (rows/columns)
- `fillna()` → replace missing values with a given value or method (e.g., forward fill)

Always inspect how much data is missing before deciding to drop or fill.

```python
import pandas as pd
import numpy as np
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, None, 35, 40, None],
    "City": ["Delhi", "Mumbai", None, "Bangalore", "Chennai"],
    "Score": [85, 91, np.nan, 88, 95]
}

b = pd.DataFrame(a)
print(b)
```

## Changing Data Types (`astype`)
- Convert a column to a different type using `astype()`.
- Examples: converting strings to numbers, integers to floats, or columns to category.

Be careful: make sure the data is compatible with the new type!

```python
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)

#1 .isna() ---> null value -> marks by True.
print(b.isna())

#2. .isnull()  --> Null valus.
print(b.isnull().sum())  # column wise null value count karega
```

```python
print(b.dropna())  # --> Remove the rows where null value
```

```python
# Filling null value of age column.
b["Age"].fillna(b["Age"].mean(),inplace = True)
print(b)

# Filling by value 0.
b["Score"].fillna(0,inplace = True)  # score me jaha null value aayegi waha zero
# se replace ho jayegi
print(b)

# Filling by placeholder("Unknown")
b["City"].fillna("Unknown",inplace = True)
print(b)
```

## Replacing Values (`replace`)
- Use `replace()` to substitute specific values or patterns.
- Useful for cleaning labels, fixing typos, or mapping categories.

```python
print(b)

b["Name"].replace("Alice","suyog",inplace=True) # yaha alice ke jagah suyog
# replace ho jayega
b["Score"].replace(88,20,inplace=True)

print(b)
```

## Dropping Rows & Columns (`drop`)
- `drop(labels, axis=0)` → drop rows by index labels.
- `drop(labels, axis=1)` → drop columns by name.
- Use `inplace=True` if you want to modify the DataFrame directly.

```python
print(b)
#Remove row
#b.drop(index =2,axis=0,inplace=True)
print(b)

# Remove column
b.drop("Score",axis = 1, inplace = True)
print(b)
```

## Detecting & Removing Duplicates
- `duplicated()` → returns a Boolean Series marking duplicate rows.
- `drop_duplicates()` → remove duplicate rows.
- You can specify `subset` (columns) and `keep` ('first', 'last', or False).

```python
data = {
    "Name": ["Alice", "Modiji", "Charlie", "Alice", "Modiji", "David"],
    "Age": [25, 30, 35, 25, 30, 40],
    "Score": [88, 92, 79, 88, 92, 85]
}

df = pd.DataFrame(data)

# .astype() --> for Data type conversion
b["Age"].astype(float) # float me convert ho jayega[.astype]
```

```python
df.duplicated().sum()  # Total Duplicate values count

df.drop_duplicates(inplace=True) # agar exact same / identical rows aati hai
#toh drop kar do
print(df)
```

```python
import pandas as pd
import numpy as np
#1). Create a messy Data Frame
data = {
    "Name" : ["Alice","Bob","Chrlie","David","Alice",None],
    "Age" : ["25","30",None,"40","25","30"],
    "City" : ["Delhi","Mumbai","Pun","Bangakore","Delhi","Mumbai"],
    "Score" : [85,None,78,88,85,91],
    "Extra" : ["x","y","z","x","y","z"]
}
df = pd.DataFrame(data) # ye ek data frame me create kar deta hai
print(df)


# 2). Handle missing values
# Fill or drop as needed [df["Name"].fillna("Unlnown name")]

# Drop{df.dropna()} --> Data frame se unn rows ko hata deta hai jisme koi v
# missing value(None) ho !
print(df.dropna())

df["Name"] = df["Name"].fillna("Abhishek")  # {.fillna} ka main work
# hai jaha "None" value hai oha value put karna apne man se koi v
df["Age"] = df["Age"].fillna("33")
df["Score"] = df["Score"].fillna(63)
print(df)


# 3. Change data types
# Convert Age to int, Score to float, etc.
df["Age"] = df["Age"].astype(int)  # {.astype ka use int,float,string ,
#, bolleanme me convert karne ke liye kiya jata hai}
df["Score"] = df["Score"].astype(float)
print(df)


#4.
# 4. Replace incorrect values
# Fix 'Pun' -> 'Pune', maybe unify Name spellings if needed
df["City"] = df["City"].replace("Pun","Pune")
print(df)

#5. Drop unnecessary rows/columns
# Remove 'Extra' or any row you don't want
df = df.drop(index= 2) # yaha mujhe kisi rows ko hatana ho to "index" ka use
# karenge
print(df)

# yedi dataframe se koi "column" ko hatana ho ho to {  df.drop(columns= ["Name of columns"]) }
df = df.drop(columns=["City"])
print(df)  # yaha columns me se city ko drop kar diya hai

# 6. Remove duplicates
# Use duplicated() and drop_duplicates()

# print(df[df.duplicated(subset=["Name"])])
df.drop_duplicates(subset=["Age"],inplace=True)
print(df) # ye Age colun ke sare duplicate value ko delete kar dega



# 7. Final clean DataFrame
# Print the cleaned DataFrame and check info()

print("\nDataFrame Info:")
df.info()

```
