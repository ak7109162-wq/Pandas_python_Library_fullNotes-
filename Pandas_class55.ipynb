# 🐼 Pandas - Class 5: Sorting & Basic Statistics
Welcome to **Class 5** of our Pandas series. Today we’ll learn how to sort data and calculate basic statistics.

## 1. Sorting Data
- Use `sort_values(by='col')` to sort by column values.
- Use `ascending=False` for descending order.
- Sort by multiple columns by passing a list.
- `sort_index()` sorts by row or column index.

```python
import pandas as pd
data = {
    "Product" : ["Pen","Notebook","Pencil","Marker","Eraser","Sharpner"],
    "Price" : [10,50,5,30,9,12],
    "Stock" : [100,60,200,80,150,120],
    "Rating" : [4.5,4.7,4.2,4.8,4.3,4.4]
}
df = pd.DataFrame(data)


#1. sort by a single column -> price ,Ascending order {by defaut ascending hota
# hai}
print(df.sort_values(by="Price"))

#2. sort by a single column -> Rating, Descending.
print(df.sort_values(by= "Rating",ascending=False))

#3. sort by multiple columns --> price and stock
df.sort_values(by=["Price","Stock"]) # uper dono wala ka combination hai

#4. sort index -> in descending order.
df.sort_index(ascending = False)



```

## 2. Descriptive Statistics
- `mean()`, `median()`, `mode()` for central tendency.
- `std()` for standard deviation.
- `describe()` for a summary of stats (count, mean, std, min, quartiles, max).

```python
print(df)

#1. Mean of numeric column only
print(df.mean(numeric_only=True)) # mean niknalne ke liye only numeric value pe lagegi

#2. Median
print(df.median(numeric_only=True))

#3. Mode
print(df.mode())

#4. standard deviation
print(df.std(numeric_only=True))

#5. describe --> complete stastistical vaues of columns.[mean,std,count,min, max]
print(df.describe())
```

## 3. Counting Values
- `value_counts()` shows the frequency of unique values in a Series.
- Use `normalize=True` to see percentages instead of counts.

```python
print(df)

#1. count how many times each product appears
print(df["Product"].value_counts())

#2. show percentage instead of counts
print(df["Product"].value_counts(normalize=True))

#3 counts values for column rating.
print(df["Rating"].value_counts())
```

## 4. Correlation & Covariance
- `corr()` computes correlation between numeric columns.
- `cov()` computes covariance.
- Correlation values range from -1 (negative) to +1 (positive).

```python
#1. correlation between all numeric column.
print(df.corr(numeric_only=True))  # ye realation batate hai

# -ve correlation -> one increases 2nd decreases
# +ve correlation -> one increase other increase.

#2. covarience between all  numeric column
print(df.cov(numeric_only=True))  # direction batate hai

#3. correlation between 2 specific column.
print(df["Price"].corr(df["Stock"])) # ye batay ga ki +ve relation hai ya
# negative
```
